```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("twitter_human_bots_dataset.csv")
columns_to_drop=['id','Unnamed: 0','created_at', 'description', 'profile_image_url', 'profile_background_image_url', 'screen_name', 'location','geo_enabled' ,'lang']
```

```python
selected_features = data.drop(columns=columns_to_drop,axis=0)
```

```python
selected_features
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
for i, feature in enumerate(selected_features.columns):
    plt.subplot(3, 4, i + 1)  # Adjust the subplot layout as needed
    sns.histplot(data=selected_features, x=feature, bins=20, kde=True)
    plt.title(f'Distribution Plot for {feature}')
    plt.xlabel(f'{feature} Values')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.savefig('density_plots.png')
plt.show()
```

```python
# Create a new DataFrame with only the 'followers_count' feature
followers_count_scaled = np.log10(selected_features['followers_count'])
followers_count_scaled.head()

plt.figure(figsize=(8, 6))
sns.histplot(data=followers_count_scaled, bins=20, kde=True)
plt.title('Distribution Plot for followers_count')
plt.xlabel('followers_count Values')
plt.ylabel('Frequency') 

# plt.show()
```
